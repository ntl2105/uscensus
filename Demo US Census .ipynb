### Data dictionnary

B01003_001E: Total Population. <br>
B02001_002E: Population, white-alone. <br>
B17001_002E: Number of persons whose income in the past 12 months is below the poverty level. <br>

B23025_005E: Number of unemployed, age 16 or older, in the civilian labor force. <br>
B23025_007E: Number of persons, age 16 or older, not in the labor force. <br>
B15003_002E: The number of persons age 25 and over who completed no schooling. <br>
B15003_017E: The number of persons age 25 and over who have a regular high school diploma. <br>
B15003_022E: The number of persons age 25 and over who hold a Bachelor's degree. <br>

Complete list of variables: https://api.census.gov/data/2012/acs5/variables.html

```python
c = Census(config.api_key) #default year is 2016, change parameter year to another year
```

```python
def getFeaturesByZipCode(variables, zip_codes):
    '''returns a dataframe with one zip code column and attributes columns'''
    data_list = []
    no_record_zip = []
    for zip_code in zip_codes :
        data_zip_code = c.acs5.zipcode((variables), zip_code)
        if len(data_zip_code) == 0:
            no_record_zip.append(zip_code)
        data_list.extend(data_zip_code)

    df = pd.DataFrame(data_list)
    print('Unable to retrieve data from US Census for {0} zip codes'.format(len(no_record_zip)))
    return df
```
